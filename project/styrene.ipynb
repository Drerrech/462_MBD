```python

#code for creating contour plot

import numpy as np
import matplotlib.pyplot as plt

# Create a grid of (x, y) values
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)

Z = X**4 + X**2 +Y**3 - y 

#plot

plt.figure(figsize=(6, 5))
contours = plt.contour(X, Y, Z, levels=10)   
plt.clabel(contours)                        
plt.title("Level Curves of f(x,y) = x + y")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.show()
```
